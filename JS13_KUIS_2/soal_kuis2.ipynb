3. Tentukan secara tepat kolom yang menjadi "label"

```python
# mengimport label encoder
from sklearn import preprocessing
# mendefinisikan label encoder
le = preprocessing.LabelEncoder()
# encode label yang ada dalam kolom 'keterkibatann_vltr'
df['keterlibatan_vltr']= le.fit_transform(df['keterlibatan_vltr'])
df['jenis_luaran_lain']= le.fit_transform(df['jenis_luaran_lain'])
df['status_luaran_lain']= le.fit_transform(df['status_luaran_lain'])
df['jenis_luaran_hki']= le.fit_transform(df['jenis_luaran_hki'])
df['luaran_hki']= le.fit_transform(df['luaran_hki'])
df['potensi_hki']= le.fit_transform(df['potensi_hki'])
df['luaran_lain']= le.fit_transform(df['luaran_lain'])
df['skema']= le.fit_transform(df['skema'])
# menggunakan Decision Tree
# dengan variabel keterlibatan volunter sampai hasil keluaran yang didapatkan
x = df.iloc[:,6:16]
y = df['tahun']
x.head()
```

4. Terapkan dua jenis metode klasifikasi dan lakukan perbandingan akurasi  dengan wajib melibatkan variabel keterlibatan volunteer sampai variabel luaran-luaran yang dihasilkan

Menggunakan decision tree

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt = DecisionTreeClassifier()

#data trainning disesuaikan ke set trainning
dt.fit(X_train, y_train)

# prediksi label set testining 
y_pred_dt = dt.predict(X_test)
print(y_pred_dt)

# set accuracy_score
acc_dt = accuracy_score(y_test,y_pred_dt)
print("set testing accuracy: {:.2f}".format(acc_dt))
print(f"set testing accuracy: {acc_dt}")
```

menggunakan SVM Polynomial dengan variable yang ditentukan

```python
from sklearn.svm import SVC

svm_poly = SVC(kernel='poly', degree=8)

# fit ke model
svm_poly.fit(X_train, y_train)

# prediksi dt
y_pred_svm_poly = svm_poly.predict(X_test)
print(y_pred_svm_poly)

# evaluasi akurasi dt
acc_svm_poly= accuracy_score(y_test, y_pred_svm_poly)

print("set testing accuracy : {:.2f}".format(acc_svm_poly))
print(f"set testing accuracy : {acc_svm_poly}")
```

5. Lakukan tunning hyperparameter berdasarkan hasil klasifikasi pada nomor 3 menggunakan metode Grid Search dan Random Search! Simpulkan parameter apa yang paling baik digunakan berdasarkan metode klasifikasi yang dipilih, dan berapa akurasinya.

hyperparameter tunning pada klasifikasi decision tree menggunakan Grid search 

```python
from sklearn.model_selection import RepeatedKFold

# inisialisasi data ke repeated k-fold
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# inisisasi model
dt = DecisionTreeClassifier()

# menamai hyperparameter yang akan digunakan
params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : list(range(5,11))
}

# inisiasi grid dari nilai  hyperparameter  dan repeated k-fold
grid = GridSearchCV(dt, param_grid=params, cv=cv)

#  %timeit untuk menghitung waktu komputasi berdasarkan grid
%timeit grid.fit(x,y)

# hasil score dengan grid.score
score = grid.score(x,y)

print(f'hasil score : {score}')
print(f'konfigurasi hyperparameter : {grid.best_params_}')
```

didapatkan hyperparameter yang dapat digunakan adalah dengan
criteria seperti ini dan max depth sebesar 5

hasil hyperparameter yang dapat digunakan adalah C
=1 , degree = 3 dan kernel yang digunakan adalah linear

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

#model SVM 
svm = SVC()

# menamai hyperparameter yang akan digunakan
params = {
    'C' : [0.1, 1, 10],
    'degree' : [3],
    'kernel' : ['linear','poly','rbf']
}

# inisiasi grid dari nilai  hyperparameter  dan repeated k-fold
grid = GridSearchCV(svm, param_grid=params, cv=cv)

# %timeit untuk menghitung waktu komputasi berdasarkan grid
%timeit grid.fit(x,y)

#hasil score dengan grid.score
score = grid.score(x,y)

print(f'hasil score: {score}')
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

Tunning hyperparameter menggunakan RandomSearch pada klasifikasi Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

# inisisasi model
dt = DecisionTreeClassifier()

#menamai hyperparameter yang akan digunakan
params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : list(range(5,11))
}

#inisiasi grid dari nilai  hyperparameter  dan repeated k-fold
grid = RandomizedSearchCV(dt, params, cv=cv)

#%timeit untuk menghitung waktu komputasi berdasarkan grid
%timeit grid.fit(x,y)

#hasil score dengan grid.score
score = grid.score(x,y)

print(f'hasil score: {score}')
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

Dengan menggunakan randomsearch hasil hyperparameter yang didapatkan berbeda
yaitu criterion sama dengan entropy dan max depthnya sebesar 8

Tunning hyperparameter menggunakan randomsearch pada klasifikasi svm polynomial




```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV

#model SVM 
svm = SVC()

#menamai hyperparameter yang akan digunakan
params = {
    'C' : [0.1, 1, 10],
    'degree' : [3],
    'kernel' : ['linear','poly','rbf']
}

# inisiasi grid dari nilai  hyperparameter  dan repeated k-fold
grid = RandomizedSearchCV(svm, params, cv=cv)

#%timeit untuk menghitung waktu komputasi berdasarkan grid
%timeit grid.fit(x,y)

#hasil score dengan grid.score
score = grid.score(x,y)

print(f'Hasil score: {score}')
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

kesimpulan, jadi hyperparameter tunning pada klasifikasi  menggunakan Grid search dan Random search memiliki hasil score akurasi yang sama yaitu hasilnya  = 1, dengan proses yang sama sama cepat tetapi yang paling baik adalah menggunakan  Grid search karena max depth yang didapatkan nilainya akan sama, berbeda dengan random search yang akan berbeda beda di setiap running nya
